```python
# Assuming 'data' is a pandas DataFrame with columns containing numerical values
# Importing stats from scipy
from scipy import stats
# List of columns you want to check for outliers
columns_to_check = ['No_of_Reassignments', 'number_cnt', 'No_of_Related_Interactions']  # Add other columns as needed

for column in columns_to_check:
    # Calculate IQR for the current column
    IQR = stats.iqr(df[column], interpolation='midpoint')
    print(f'IQR for {column}: {IQR:.2f}')
    
    # Calculate Q1 and Q3
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    print(f'Q1 for {column}: {Q1:.2f}')
    print(f'Q3 for {column}: {Q3:.2f}')
    
    # Calculate lower and upper limits
    low_limit = Q1 - 1.5 * IQR
    upp_limit = Q3 + 1.5 * IQR
    print(f'Low limit for {column}: {low_limit:.2f}')
    print(f'Upper limit for {column}: {upp_limit:.2f}')
    
    # Optionally, you can handle outliers by filtering the data
    # For example, removing the rows with outliers
    df = df[(df[column] >= low_limit) & (df[column] <= upp_limit)]

# After the loop, 'data' will have rows with outliers removed for specified columns.
```

```python
df
```

```python
# Select only the numeric columns in the dataset
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for column in numeric_cols:
    IQR = stats.iqr(df[column], interpolation='midpoint')
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    low_limit = Q1 - 1.5 * IQR
    upp_limit = Q3 + 1.5 * IQR

    # Replace outliers with NaN
    df[column] = np.where((df[column] < low_limit) | (df[column] > upp_limit), np.nan, df[column])
```

```python
df
```

```python
# Define a threshold below which a category is considered "rare"
threshold = 0.01  # 1% of the data

for column in df.select_dtypes(include=['object']).columns:  # Loop through categorical columns
    value_counts = df[column].value_counts(normalize=True)  # Get value counts as proportions
    
    # Replace rare categories with 'Other'
    rare_categories = value_counts[value_counts < threshold].index
    df[column] = df[column].replace(rare_categories, 'Other')

    print(f"Rare categories in {column} replaced with 'Other'")
```

```python
from sklearn.metrics import classification_report, f1_score
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Initialize Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Train the model
model.fit(x_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(x_test)
```

```python
# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')

# Print classification report for more detailed metrics
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix to understand misclassifications
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for tuning
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5)

# Fit GridSearchCV
grid_search.fit(x_train, y_train)

# Print best parameters and score
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best Score: {grid_search.best_score_}")
```
